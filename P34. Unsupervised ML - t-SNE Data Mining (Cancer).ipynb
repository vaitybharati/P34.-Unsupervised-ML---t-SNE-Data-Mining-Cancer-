# Unsupervised ML - t-SNE Data Mining 

```python
# Import Libraries
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
from bioinfokit.visuz import cluster
```

```python
# Import Dataset
cancer=pd.read_csv('Database/TSNE_data.csv')
cancer
```

```python
# Convert data to array format
cancer_ary=cancer.values
cancer_ary
```

```python
# separate array into input and output components
X=cancer_ary[:,1:]
Y=cancer_ary[:,0]
```

## TSNE implementation

```python
tsne=TSNE(n_components=2)
cancer_tsne=tsne.fit_transform(X)
cancer_tsne
```

```python
# Cluster Visualization (Plot will be stored in the default directory)
color_class=Y
cluster.tsneplot(score=cancer_tsne, colorlist=color_class, legendpos='upper right',legendanchor=(1.15, 1))
```
